```python
import gensim
from gensim.summarization import summarize
text = open("C:/Users/arjun/Documents/Online Assignments/RR/train.csv", "r")

# joining with space content of text
text = ' '.join([i for i in text]) 

# replacing ',' by space
text = text.replace(",", " ")  

#displaying result
#print(text)
len(text.split()) 

# find total words
text.split()

#summarize(text)
word_count = summarize(text,word_count=8000)
len(word_count)


print("Word count summary - ")
print(word_count)
```

```python
# We can also ask to summarise at the number of words rather than ratio

# Summary (0.5% of the original content).

summ_per = summarize(text, ratio = 0.05)
print("Percent summary - ")
print(summ_per)
```
